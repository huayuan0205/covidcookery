```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Reddit_Scraper_Covidcookery.csv")
```

```python
import spacy
```

```python
nlp = spacy.load("en_core_web_md")
foods = nlp("orange milk flour meat bread pepper carrot cake cheese")
```

```python
nlp(df["title"][0])
```

```python
import warnings
warnings.filterwarnings("ignore", message=r"\[W008\]", category=UserWarning)
```

```python
lists = []
for n in df["title"]:
    line_nlp = nlp(n)
    parts = ""
    for chunk in line_nlp.noun_chunks:
        final_chunk = ""
        for token in chunk:
            if token.pos_ == 'NOUN' or token.pos_ == 'PROPN':
                sim = token.similarity(foods)
                if sim > 0.6:
                    final_chunk = final_chunk + token.lemma_ + " "
        if final_chunk:
            parts = parts + final_chunk.strip() + ","
    lists.append(parts)
```

```python
new_list = []
for n in lists:
    n = n.rstrip(",")
    n = n.rstrip(" ")
    new_list.append(n)
```

```python
new_list
```

```python
df_food = pd.DataFrame(new_list, columns=["food keywords"])
df_final = df.join(df_food)
df_final.to_csv("Covidcookery_keywords.csv")
```
